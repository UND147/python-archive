```python
scores = [120, 150, 180, 200, 170]
maxScore = scores[0]
for i in range(1, len(scores)):
    if maxScore < scores[i]:
        maxScore = scores[i]
print(maxScore)
```

```python
scores = [120, 150, 180, 200, 170]
maxScore = scores[0]

for s in scores[1:]:  
    if maxScore < s:
        maxScore = s
print(max(scores))
# print(maxScore)
print(f"최고점수: {maxScore}")
```

```python
rooms = [
    [3,1,2],
    [2,0,1],
    [1,3,2]
]
total = 0
for row in rooms:
    for item in row:
        if item != 0:
            total += item
print(f"총 수집한 아이템 수: {total}")
```

```python
del sum
rooms = [
    [3, 1, 2],
    [2, 0, 1],
    [1, 3, 2]
]
total = 0

for row in rooms:
    total += sum(row)

print(f"총 수집한 아이템 수: {total}")
```
