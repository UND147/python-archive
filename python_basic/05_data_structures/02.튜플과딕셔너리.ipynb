```python
stock_prices = {
    '삼성전자':[81000,81500,82000],
    'SK하이닉스':[140000,141000,139500],
    '네이버':[350000,355000,345000]
}
total1 = total2 = total3 = 0
for i in stock_prices['삼성전자']:
    total1 += i
total1 /= 3
print(f"#삼성전자 평균 주식 가격:{int(total1)}")
for i in stock_prices['SK하이닉스']:
    total2 += i
total2 /= 3
print(f"#SK하이닉스 평균 주식 가격:{int(total2)}")
for i in stock_prices['네이버']:
    total3 += i
total3 /= 3
print(f"#네이버 평균 주식 가격:{int(total3)}")
```

```python
stocks_info = {
    '삼성전자':{'최고가':85000,'최저가':80000,'현재가':82000},
    'SK하이닉스':{'최고가':145000,'최저가':139000,'현재가':140500},
    '네이버':{'최고가':360000,'최저가':340000,'현재가':350000}
}
for key,value in stocks_info.items():
    max_value = value['최고가']
    min_value = value['최저가']
    print(f"{key}-최고가:{max_value},최저가:{min_value}")
```

```python
stock_prices = {
    '삼성전자':[81000,81500,82000],
    'SK하이닉스':[140000,141000,139500],
    '네이버':[350000,355000,345000]
}
for key, value in stock_prices.items():
    print(value)
    print(sum(value))
    avg = sum(value) / len(value)
    print(f"#{key} 평균 주식가격: {int(avg)}")
```
