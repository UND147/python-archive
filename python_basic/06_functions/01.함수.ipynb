```python
def easy_recipe(a,b):
    print(f"{a}와{b}를 이용해서 맛있는 요리를 시작합니다")
    result = a + b + '볶음밥'
    print('요리완료!')
    return result

easy_recipe('치즈', '불닭')
```

```python
def cal_kda(kill, death, assist):
    kda = (kill+assist)/death
    return kda
cal_kda(10,2,7)
```

```python
def cal_average_gold(a, b, c):
    avg = (a+b+c)/3
    return avg
cal_average_gold(12000,15000,18000)
```
